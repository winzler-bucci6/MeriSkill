# Data Visualisations

```python
#Correlation map for all numeric values
corr=data.corr()
plt.figure(figsize=(12,8))
sns.heatmap(corr,annot=True, cmap='coolwarm',fmt=".2f")
plt.title('Correlation Map')
plt.show()
```

#Insights:
Clearly Job level is strongly correlated to Monthly Income ie Monthly income increases as the Job level increases.
There also seems to be some correlation between the total number of working years with Job Level and Montlhy Income.

```python
data.columns
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x='OverTime',data=data)
plt.title('Overtime Count')
plt.show()
```

#Insights:
A lot of people do not work overtime

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='OverTime',y='Age',data=data)
plt.title('Relation between Overtime and Age')
plt.show()
```

```python
gender_distribution = data.groupby(['Department', 'Gender']).size().unstack(fill_value=0)
gender_distribution
```

```python
plt.figure(figsize=(12,6))

gender_distribution.plot(kind='bar', stacked='True')

plt.xlabel('Department')
plt.ylabel('Count')
plt.title('Gender Distribution by Department')
plt.xticks(rotation=45)
plt.legend(title='Gender')

plt.tight_layout()
plt.show()
```

#Insights: There are more males than females in all departments.

```python
plt.figure(figsize=(10, 6))

sns.countplot(x='JobRole', hue='BusinessTravel', data=data, palette='Set2')

plt.title('Relation between Job Role and Business Travel')
plt.xlabel('Job Role')
plt.ylabel('Count')
plt.xticks(rotation=90) 
plt.legend(title='Business Travel', loc='upper right')

plt.tight_layout()
plt.show()
```

#Insights:The job roles of Sales Executive,Research Scientist and Laboratory Technician travel frequently as compared to other roles.The non-travel roles are Human Resources,Research Director and Sales Representative.

```python
plt.figure(figsize=(10, 6))

sns.barplot(x='EducationLevel', y='TotalWorkingYears', data=data, palette='Set2')

plt.title('Number of Years Worked by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Number of Years Worked')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability

plt.tight_layout()
plt.show()
```

#Insights:In terms of how education levels relate to Number of Years Worked,it is clear that the higher the education level the more the number of years worked

```python
attrition_counts = data['Attrition'].value_counts()
plt.pie(attrition_counts, labels=attrition_counts.index, autopct='%1.1f%%')
plt.title('Employee Attrition')
plt.axis('equal') 
plt.show()
```

#Insights:A small percentage of employees have left the company.

```python
#Correlation visualisation for those employees who have left

Left_Employees=data[data['Attrition']=='Yes']

correlation_matrix = Left_Employees.corr()
plt.figure(figsize=(12, 8)) 
heatmap = sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm',fmt=".2f")
plt.title('Correlation Heatmap for Employees who left')
plt.show()
```
